```python
import pikepdf
import os

def recover_pdf(input_folder, output_folder):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(".pdf"):
            print(f"Processing: {filename}...")
            try:
                # Opening the PDF with pikepdf often strips away 
                # simple JS-based 'expiry' restrictions automatically
                with pikepdf.open(os.path.join(input_folder, filename)) as pdf:
                    # Save a copy without the restrictions
                    pdf.save(os.path.join(output_folder, f"recovered_{filename}"))
                print(f"Successfully recovered: {filename}")
            except Exception as e:
                print(f"Could not recover {filename}: {e}")

# Usage
recover_pdf(r"C:\Users\SamuelJoshuaRaj\OneDrive - CYGNUSA Technologies\Desktop\merged bodinayakanur",r"C:\Users\SamuelJoshuaRaj\OneDrive - CYGNUSA Technologies\Desktop\recovered" )
```
