```python
import os
from PyPDF2 import PdfMerger

input_folder = r"C:\Users\SamuelJoshuaRaj\OneDrive - CYGNUSA Technologies\Desktop\Bodinayakanur pdf sorted"
output_folder = r"C:\Users\SamuelJoshuaRaj\OneDrive - CYGNUSA Technologies\Desktop\merged bodinayakanur"

os.makedirs(output_folder, exist_ok=True)

pdf_files = sorted([f for f in os.listdir(input_folder) if f.endswith(".pdf")])

total_files = len(pdf_files)
parts = 10

files_per_part = total_files // parts
extra = total_files % parts

start = 0

for i in range(parts):
    merger = PdfMerger()

    end = start + files_per_part + (1 if i < extra else 0)

    for file in pdf_files[start:end]:
        merger.append(os.path.join(input_folder, file))

    output_file = os.path.join(output_folder, f"merged_part_{i+1}.pdf")

    with open(output_file, "wb") as f:
        merger.write(f)

    merger.close()
    print(f"Created: merged_part_{i+1}.pdf")

    start = end

print("Done ✅ No quality loss at all.")
```
