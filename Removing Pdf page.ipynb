```python
import os
from PyPDF2 import PdfReader, PdfWriter

# Input and Output folders
input_folder = r"C:\Users\SamuelJoshuaRaj\Downloads\200-eroll"
output_folder = r"C:\Users\SamuelJoshuaRaj\OneDrive - CYGNUSA Technologies\Desktop\Bodinayakanur pdf sorted"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

for file_name in os.listdir(input_folder):
    if file_name.endswith(".pdf"):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name)

        reader = PdfReader(input_path)
        writer = PdfWriter()

        total_pages = len(reader.pages)

        # Skip files with too few pages
        if total_pages <= 4:
            print(f"Skipping {file_name} (not enough pages)")
            continue

        # Keep pages except first 3 and last 1
        for i in range(3, total_pages - 1):
            writer.add_page(reader.pages[i])

        # Save output
        with open(output_path, "wb") as f:
            writer.write(f)

        print(f"Processed: {file_name}")

print("All files processed successfully!")
```
